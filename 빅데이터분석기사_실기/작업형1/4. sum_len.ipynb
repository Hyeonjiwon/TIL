```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Name': ['합격핑', '만점핑', '열심핑', '도전핑', '재수핑'],
    'Age': [24, 27, 22, 32, 29],
    'Score': [85, 90, 78, np.nan, 95],
})
df
```

```python
cond = df['Age'] > 25
cond
```

```python
# sum : 총 합계
# len : 행 길이
print("sum:", sum(cond)) # F->0 , T->1
print("len:", len(cond))
```

```python
# len으로 true의 수를 찾는 방법 -> true만 선택된 데이터 프레임을 넣어준다.
df[cond]
```

```python
len(df[cond])
```

```python
print(len(df[cond]))
print(len(df[df['Age']>25]))
```

```python
# count()는 결측치를 체크하지 않음
df[cond].count()
```
