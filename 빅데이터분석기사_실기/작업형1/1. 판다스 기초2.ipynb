### 데이터 추가
- 결측값으로 추가(NaN 대입)
- 리스트 형태로 데이터(행) 추가
- 딕셔너리 형태로 데이터(행) 추가

```python
# 결측값으로 추가
# 원두 컬럼을 만들고 결측값(NaN)으로 대입
import numpy as np
df['원두'] = np.nan
df
```

```python
# 아메리카노의 원두는 '콜롬비아' 데이터 추가
df.loc[0, '원두'] = '콜롬비아'
df
```

```python
# 리스트 형태로 데이터(행) 추가
df.loc['시즌'] = ['크리스마스라떼', 6000, 0, 500, '한국']
df
```

```python
# 딕셔너리 형태로 데이터(행) 추가
# 키 값에 매칭 시켜 데이터를 넣기 때문에 컬럼 수에 맞추지 않아도 데이터 추가 가능
df.loc[7] = {
    '메뉴': '에스프레소',
    '가격': 2000,
    '칼로리': 10
}
df
```

### 정렬 Sort
- default 내림차순 ascendig=True
- 인덱스 기준
- 값 기준
- 인덱스 새로 만들기

```python
df = df.drop('시즌', axis=0)
df
```

```python
# 인덱스 기준 - default 내림차순
df.sort_index()
```

```python
# 인덱스 기준 - 오름차순
df.sort_index(ascending=False)
```

```python
# 값 기준 - 오름차순
# 가격 기준 정렬
df.sort_values('가격')
```

```python
# 값 기준 - 내림차순
# 가격 기준 정렬
df.sort_values('가격', ascending=False)
```

```python
# 값 기준 - 내림차순
# 메뉴 기준 정렬
df.sort_values('메뉴')
```

```python
# 기준 n개 정렬
# 가격과 메뉴 기준 정렬
# 가격 - 내림차순 , 메뉴 - 오름차순
df = df.sort_values(['가격', '메뉴'], ascending=[False, True])
df
```

```python
# 인덱스 새로 만들기
# 기존 인덱스 값 삭제 : drop = True
df.reset_index(drop=True)
```
