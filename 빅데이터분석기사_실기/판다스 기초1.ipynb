### 자료형 변환

```python
data = {
    "메뉴":['아메리카노', '카페라떼', '카페모카', '바닐라콜드브루'],
    "가격":[4100, 4600, 4600, 5100],
    "할인율":['0.5', '0.1', '0.2', '0.3'], # object
    "칼로리":[10,180,420,320],
}
df = pd.DataFrame(data)
df.info()
```

```python
# 자료형 변환 / astype / object -> float
df['할인율'].astype('float')
```

```python
df['할인율'] = df['할인율'].astype('float')
```

```python
df.info()
```

### 새로운 컬럼 추가

```python
df
```

```python
# 새로운 컬럼 추가
df['new'] = 0
df
```

```python
# 할인가 컬럼 추가
df['할인가'] = df['가격'] * (1-df['할인율'])
df
```

```python
# 결측값(NaN)으로 추가
import numpy as np
df['원두'] = np.nan
df
```

### 데이터 삭제
- axis = 1 : 컬럼 삭제
- axis = 0 : 행 삭제

```python
# 컬럼 삭제 - drop('컬럼명', axis=1)
df = df.drop('new', axis=1)
df
```

```python
# 행 삭제 - drop('index 번호', axis=0)
df = df.drop(1, axis=0)
df
```

### CSV 저장하기

```python
# csv로 저장하기 - to_csv('파일명.csv', index=False)
df.to_csv('data2.csv', index=False)
```
