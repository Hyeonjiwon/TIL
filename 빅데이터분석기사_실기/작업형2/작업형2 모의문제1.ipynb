```python
import pandas as pd

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
y_test = pd.read_csv('y_test.csv')

# EDA
# 결측치 - 없음
# print(train.isnull().sum())
# print(test.isnull().sum())

# 이상치
# print(train.describe(include='O'))
# print(test.describe(include='O'))

cols = train.select_dtypes(include='O').columns
# print(cols)

# for col in cols:
#     print(train[col].unique())
#     print(test[col].unique())

# 데이터 전처리
# 카테고리 데이터 - 라벨인코딩
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

# print(test.head(1))

for col in cols:
    train[col] = encoder.fit_transform(train[col])
    test[col] = encoder.transform(test[col])

print(test.head(1))

# 원핫인코딩
# all_df = pd.concat([train[cols], test[cols]])
# all_df = pd.get_dummies(all_df)
# print(all_df.head(1))

# train = pd.get_dummies(train)
# test = pd.get_dummies(test)
# print(train.shape)
# print(test.shape)


# train/validation 나누기
from sklearn.model_selection import train_test_split

# print(train.info())
y = train.pop('Attrition_Flag')
# print(train.info())
X_tr, X_val, y_tr, y_val = train_test_split(train, y, test_size=0.1, random_state=123)

# 모델 선택 -> 학습 -> 예측
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_tr, y_tr)
pred = model.predict_proba(X_val)

# print(pred[:10])
# print(pred[:,1])

# 평가 지표
from sklearn.metrics import roc_auc_score

score = roc_auc_score(y_val, pred[:, 1])

# LabelEncoder 0.9886321775567528
print(score)

# CLIENTNUM,Attrition_Flag
# 788544108,0.633
# 719356008,0.123
# 712142733,0.355

# 결과 내보내기
pred = model.predict_proba(test)
import numpy as np
submit = pd.DataFrame({
        'CLIENTNUM': test['CLIENTNUM'],
        'Attrition_Flag': np.round(pred[:, 1],3)
})

print(submit.shape)

submit.to_csv("0000.csv", index=False)
```

## 검증 데이터 분리

```python
from sklearn.model_selection import train_test_split

y = train.pop('Attrition_Flag')
print()

X_tr, X_val, y_tr, y_val = train_test_split(train, y, test_size=0.2, random_state=123)
```

```python
X_tr.head(2)
```

## 모델 & 평가

```python
# 모델 불러오기
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=123)

# 학습
model.fit(X_tr, y_tr)

# 예측
pred = model.predict(X_val)
```

```python
pred
```

```python
# 자체 평가
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# 정확도
print(accuracy_score(y_val, pred))

# 정밀도
print(precision_score(y_val, pred))

# 재현율 (민감도)
print(recall_score(y_val, pred))

# F1
print(f1_score(y_val , pred))

# baseline
# 0.9666872301048736
# 0.9464285714285714
# 0.8346456692913385
# 0.8870292887029287

# label
# 0.9642196175200494
# 0.9414414414414415
# 0.8228346456692913
# 0.8781512605042017

# one-hot
# 0.9574336829117829
# 0.9223744292237442
# 0.7952755905511811
# 0.8541226215644819
```

```python
# roc-auc
pred_proba = model.predict_proba(X_val)

print(roc_auc_score(y_val, pred_proba[:,1])) # 이탈할 고객의 확률

# baseline
# 0.9877843314574706

# label
# 0.9888326642051968

# one-hot
# 0.986622237326406
```
