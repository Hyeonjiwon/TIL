## 라이브러리 및 데이터 불러오기
### 데이터: 성인 인구조사 소득 예측
- id : 고유의 id 값
- age : 나이
- workclass : 고용 형태
- fnlwgt : 사람 대표성을 나타내는 가중치 (final weight의 약자)
- education : 교육 수준
- education_num : 교육 수준 수치
- marital_status: 결혼 상태
- occupation : 업종
- relationship : 가족 관계
- race : 인종
- sex : 성별
- capital_gain : 양도 소득
- capital_loss : 양도 손실
- hours_per_week : 주당 근무 시간
- native_country : 국적
- income : 수익(예측해야 하는 값)
    - "> 50K" : 1
    - "<= 50K" : 0

```python
# 데이터 불러오기
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
# 데이터 크기
print(train.shape)
print(test.shape)
```

```python
train.shape, test.shape
# train 컬럼 -> 16개
# test 컬럼 -> 15개
# head로 살펴보면 test 데이터에 income 컬럼이 없음
```

```python
print(test.head())
```

```python
# 예측 해야할 값
train['income'].value_counts()
```

```python
# 조건
cond0 = train['income'] == '<=50K' # 0 하위 소득
cond1 = train['income'] == '>50K' # 1 상위 소득

cond_female = train['sex'] == 'Female'
cond_male = train['sex'] == 'Male'
```

```python
# 남성과 여성의 수
len(train[cond_female]), len(train[cond_male])
```

```python
# 남성 중 0과 1 (인원 수)
len(train[cond_male & cond0]), len(train[cond_male & cond1])
```

```python
# 남성 중 0과 1 (비율)
total = train[cond_male]

len(train[cond_male & cond0])/len(total), len(train[cond_male & cond1])/len(total)
```

```python
# 여성 중 0과 1 (인원 수)
len(train[cond_female & cond0]), len(train[cond_female & cond1])
```

```python
# 여성 중 0과 1 (비율)
total = train[cond_female]

len(train[cond_female & cond0])/len(total), len(train[cond_female & cond1])/len(total)
```
